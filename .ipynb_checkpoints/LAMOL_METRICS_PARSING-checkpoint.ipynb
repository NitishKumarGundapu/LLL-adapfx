### Nine Epochs Accuracy

```python
metrics_dir_lll = r"/home/ai20mtech14009/surbhi/Shadab/Final/Lamol_Nine_Epochs_Without_Adapters/models/gpt2/lll/sst_srl_woz.en_0.2/sst"
metrics_dir_finetune = r"/home/ai20mtech14009/surbhi/Shadab/Final/Lamol_Nine_Epochs_Without_Adapters/models/gpt2/finetune/sst_srl_woz.en/sst"
```

```python
from collections import defaultdict
with open(os.path.join(metrics_dir_lll, "metrics.json")) as file_:
    metrics_json = json.load(file_)
    
task_metrices_dict = defaultdict(list)
for each_metric in metrics_json:
    for each_key, each_val in each_metric.items():
        task_metrices_dict[each_key].append(each_val)

tasks_df = []
column_names = []
for each_metric, result in task_metrices_dict.items():
    tasks_df.append(pd.DataFrame(result))
    fields = [(each_metric,each_key_result) for each_key_result in result[0].keys()]
    column_names.extend(fields)
# last_epoch_metrics = metrics_json[-1]
# last_epoch_metrics["woz.en"]

concatenated_df_normal_nine_epochs = pd.concat(tasks_df, axis=1)

concatenated_df_normal_nine_epochs.columns = pd.MultiIndex.from_tuples(column_names)
concatenated_df_normal_nine_epochs.index = ["Epoch_" + str(each_val) for each_val in range(1,10)]
```
